```python
#6. To find all the movies that are released in one language
def get_unique_values(country_list):
    '''
    The function takes the genre list returns a list of all the total unique countries, number of movies
    that have more than one country of releases and number of movies having only one country of release
    '''
    more_than_one = 0
    only_one = 0
    unique_country = []
    for country in country_list:
        try:
            names = country.split(",")
            if len(names) > 1:
                more_than_one += 1
            elif len(names) == 1:
                only_one += 1
        except:
            pass
        for country in names:
            if country not in unique_country:
                unique_country.append(country)
    
    return unique_country, more_than_one, only_one
```

```python

total_unique_countrylist, more_than_one_country, only_one_country = get_unique_values(df['country'].unique())
# removing the blank country added to the list
total_unique_countrylist.remove('')
print('Total Number of Unique countries  are: ', len(total_unique_countrylist))
print('Movies having more than one country of release: ', more_than_one_country)
print('Movies having only one country of release: ', only_one_country)
```

```python
country_dict = {}

for val in total_unique_countrylist:
    country_dict[val] = 0
```

```python
temp_df = df[df['country'].notna()]
```

```python
for country in total_unique_countrylist:
    count=temp_df[temp_df['country'].str.contains(country)].shape[0]
    country_dict[country] = count
```

```python
country_movie_count = pd.DataFrame(columns=['Country', 'Movie Count'], 
                      data = {'Country':[val for val in country_dict.keys()],
                     'Movie Count': [val for val in country_dict.values()]}).sort_values(by='Movie Count', ascending=False).reset_index(drop=True)
```

```python
# Top 10 countries with most movie releases
plt.figure(figsize=(12,10))
plt.grid(axis='x',color='black', linestyle = ':', alpha=0.5)
plt.title('Top 10 Countries with most movie releases', fontname='monospace', fontsize=25, y=1.05)
a=sns.barplot(x='Movie Count', y='Country', data=country_movie_count[:10], palette='rocket')
genres = country_movie_count['Country'][:10].tolist()
for i, val in enumerate(genres):
    x_val = country_movie_count[country_movie_count['Country'] == val]['Movie Count'].values[0]
    a.text(y=i, x= x_val -130, 
           s=str(x_val),
          fontsize=14, fontname='monospace', color='white')

plt.xlabel('Count', fontsize=15)
plt.ylabel('Country', fontsize=15)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.show()
```

```python
def get_top10_directors(df,cont):
        df1 = df[df["type"] == "Movie"]
        df2 = df1[df1["country"] == cont]
        df3=df2[df2['country'].notna()]
        directors = ",".join(df3['director'].fillna('')).split(',')
        unique_director=list(set(directors))
        unique_director.remove('')
        unique_director_dict = {}
        for val in unique_director_dict:
            unique_director_dict[val] = 0
        for i in unique_director:
            ser =df3['director']
            count=0
            for index, value in ser.items():
                value1=str(value).split(',')
                for j in value1:
                    if j==i:
                        count=count+1
                        unique_director_dict[i] = count
        return unique_director_dict
    
    
```

```python

def get_dataframe_country(unique_director_dict):
        director_count = pd.DataFrame(columns=['Directors', 'Count'], 
              data = {'Directors':[val for val in unique_director_dict.keys()],
            'Count': [val for val in unique_director_dict.values()]}).sort_values(by='Count', ascending=False).reset_index(drop=True)
        return director_count[:10]
```

```python
India_dir=get_top10_directors(df,"India")
df_India_dir=get_dataframe_country(India_dir)
UnitedStates_dir=get_top10_directors(df,"United States")
df_UnitedStates_dir=get_dataframe_country(UnitedStates_dir)
UnitedKingdom_dir=get_top10_directors(df,"United Kingdom")
df_UnitedKingdom_dir=get_dataframe_country(UnitedKingdom_dir)
```

```python
#9
fig, axes = plt.subplots(3,1,figsize=(20,25))
sns.barplot(ax=axes[0],x='Count', y='Directors', data=df_India_dir,palette='Blues_r')
sns.barplot(ax=axes[1],x='Count', y='Directors', data=df_UnitedStates_dir,palette='Blues_r')
sns.barplot(ax=axes[2],x='Count', y='Directors', data=df_UnitedKingdom_dir,palette='Blues_r')
axes[0].title.set_text('Top directors of India')
axes[1].title.set_text('Top directors of USA')
axes[2].title.set_text('Top directors of United Kingdom')
plt.show()    
```

```python
#10. Top 10 actors on Netflix with most movies
def get_top10_actors(df,cont):
        df1 = df[df["type"] == "Movie"]
        df2 = df1[df1["country"] == cont]
        df3=df2[df2['country'].notna()]
        actors = ",".join(df3['cast'].fillna('')).split(',')
        unique_actor=list(set(actors))
        unique_actor.remove('')
        unique_actor_dict = {}
        for val in unique_actor_dict:
            unique_actor_dict[val] = 0
        for i in unique_actor:
            ser =df3['cast']
            count=0
            for index, value in ser.items():
                value1=str(value).split(',')
                for j in value1:
                    
                    if j==i:
                        count=count+1
                        unique_actor_dict[i] = count
        return unique_actor_dict
```

```python
def get_dataframe_actor(unique_actor_dict):
        actor_count = pd.DataFrame(columns=['Actors', 'Count'], 
              data = {'Actors':[val for val in unique_actor_dict.keys()],
            'Count': [val for val in unique_actor_dict.values()]}).sort_values(by='Count', ascending=False).reset_index(drop=True)
        return actor_count[:10]
```

```python
India_act=get_top10_actors(df,"India")
df_India_act=get_dataframe_actor(India_act)
UnitedStates_act=get_top10_actors(df,"United States")
df_UnitedStates_act=get_dataframe_actor(UnitedStates_act)
UnitedKingdom_act=get_top10_actors(df,"United Kingdom")
df_UnitedKingdom_act=get_dataframe_actor(UnitedKingdom_act)
```

```python
fig, axes = plt.subplots(3,1,figsize=(20,25))
sns.barplot(ax=axes[0],x='Count', y='Actors', data=df_India_act,palette='Blues_r')
sns.barplot(ax=axes[1],x='Count', y='Actors', data=df_UnitedStates_act,palette='Blues_r')
sns.barplot(ax=axes[2],x='Count', y='Actors', data=df_UnitedKingdom_act,palette='Blues_r')
axes[0].title.set_text('Top Actors of India')
axes[1].title.set_text('Top Actors of USA')
axes[2].title.set_text('Top Actors of United Kingdom')
plt.show()    
```
